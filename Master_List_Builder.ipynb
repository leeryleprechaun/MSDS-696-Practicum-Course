A program to combine all of the strain lists that were created into a single global list

```python
# Loadning in packages
import numpy
import pandas as pd
```

```python
# Loadin in the first table
propagate = pd.read_csv('data/Propagate_Lab_List.csv')
propagate
```

There are columns that need to be removed, would also like the format to be - Strain Number, Strain Name, Supplier, Attenuation, Temperature, Flocculation, Alcohol Tolerance, Sta1+, Description

```python
# Printing column names
print(propagate.columns.tolist())
```

```python
# dropping unwanted columns
propagate = propagate.drop(columns = ['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 10', 'Unnamed: 11'] , axis = 0)
propagate
```

```python
# Adding in Supplier
propagate['Supplier'] = 'Propagate Lab'
propagate
```

```python
# Re-Ordering the columns
propagate = propagate[['Strain', 'Name', 'Supplier', 'Attenuation',
                       'Flocculation', 'Fermentation_Temperature', 'Alcohol_Tolerance', 'STA+', 'Description']]
propagate
```

Loading in Yeast from another supplier to attach to the Propagate Sheet

```python
# Loading in White Labs Data
whitelabs = pd.read_csv('data/White_Labs_Strains_with_number.csv')
whitelabs
```

```python
# Adding Supplier Column
whitelabs['Supplier'] ='White Labs'
# Renaming Strain to name
whitelabs = whitelabs.rename(columns={'Strain' : 'Name'})
whitelabs
```

```python
# Renaming Number to Strain
whitelabs = whitelabs.rename(columns={'Number' : 'Strain', 'STA1' : 'STA+'})
whitelabs
```

```python
# Adding the lists together
full_frame = pd.merge(propagate, whitelabs, how='outer')
full_frame
```

Adding in the Wyest Data

```python
# Loading in Wyeast Data
wyeast = pd.read_csv('data/wyeast_strains.csv')
wyeast
```

```python
# Adding Supplier Column
wyeast['Supplier'] ='Wyeast'
wyeast['STA+'] = 'Not Listed'
# Renaming Strain to name
wyeast = wyeast.rename(columns={'Number' : 'Name'})
wyeast
```

```python
# Adding the lists together
df = pd.merge(propagate, whitelabs, how='outer')
df
```

```python
df2 = pd.merge(df, wyeast, how='outer')
df2
```

Adding in the Omega Yeast Strains

```python
# Loading in Omega Data
omega = pd.read_csv('data/Omega_Strains2024.csv')
omega
```

```python
# Adding Supplier Column
omega['Supplier'] ='Omega'
# Renaming Strain to name
omega = omega.rename(columns={'Number' : 'Name'})
omega
```

```python
#Merging all the frames together
df3 = pd.merge(df2, omega, how='outer')
df3
```

Load in the Imperial Yeast Strain Library

```python
# Loading in Imperial Data
imperial = pd.read_csv('data/imperial_2024.csv')
imperial
```

```python
# Adding Supplier Column
imperial['Supplier'] ='Imperial'
# Renaming Strain to name
imperial = imperial.rename(columns={'Number' : 'Strain'})
imperial
```

```python
#Merging all the frames together
df4 = pd.merge(df3, imperial, how='outer')
df4
```

Loading in the Inland Island Strain

```python
# Loading in Inland Island Data
inland = pd.read_csv('data/inlandisland_strains.csv')
inland
```

```python
# Adding Supplier Column
inland['Supplier'] ='Inland Island'
inland
```

```python
#Merging all the frames together
df5 = pd.merge(df4, imperial, how='outer')
df5
```

```python
#save the final list
df5.to_csv('data/global_yeast_List.csv')
```
