```python
# Verifying that the page request is valid
from gazpacho import get, Soup

url = 'https://inlandislandyeast.com/yeast-library/inis-001-northern-california-ale-yeast/'
# fetch step
html = get(url)

# extract step
soup = Soup(html)
```

```python
# use find() to parse a specific HTML
gaz_data = soup.find('div', {'class':'container'})
gaz_data
```

### Pull The Strain Name

```python
# Pull The Strain Name

def pull_strain_number(soup):
    strain_name = soup.find('div', {'class':'container'})
    x = str(strain_name)
    start = x.find('INIS')
    strain = x[start:start+8]
    # print(strain)
    return strain
```

```python
pull_strain_number(soup)
```

### Pull The Strain Number

```python
# Pull The Strain Number

def pull_strain_name(soup):
    strain_number = soup.find('div', {'class':'container'})
    x = str(strain_number)
    start = x.find('</h1></div></div>	')
    number = x[start-20:start]
    # print(number)
    return number
```

```python
pull_strain_name(soup)
```

### Pull The Description

```python
# Pull and save the description

def pull_strain_description(soup):
    description = soup.find('p')
    #print(description)
    x= str(description)
    start = x.find('<p class="jet-listing-dynamic-field__content">')
    end = x.find('</p>, <p>Category</p>, <p>Attenuation')
    description = x[start+46:end]
    # print(description)
    return description
```

```python
pull_strain_description(soup)
```

### Get Stats on Strain

```python
# use find() to parse a specific HTML
gaz_data = soup.find('div', {'class':'container'})
gaz_data
```

```python
def pull_stats(soup):
    raw = soup.find('div', {'class':'jet-listing-dynamic-field'})
    x= str(raw)
    # print(x)
    fs = x.find('div')
    flocculation = x[fs+3246:fs+3255]
    print(flocculation)

    attens = x.find('b3a09c4')
    attenuation = x[attens+397:attens+402]
    # print(attenuation)

    abve = x.find('b13a83c')
    abv = x[abve+397:abve+402]
    # print(abv)

    temps = x.find('7ea77a5')
    temp = x[abve+397:abve+402]
    # print(temp)

    return(flocculation, attenuation, abv, temp)
```

```python
pull_stats(soup)
```

### Pull all the urls

```python
import pandas as pd
url = pd.read_csv('data/inlandisland.csv')
```

```python
url_list = url['Address'].to_list()
```

```python
url_list
```

```python
import pandas as pd

def loadTable(soup):
    strain = pull_strain_name(soup)
    #pullMetrics(soup)
    number = pull_strain_number(soup)
    flocculation, attenuation, abv, temp = pull_stats(soup)
    description = pull_strain_description(soup)
    entry = {'Strain': [strain],
             'Number' : [number],
        'Attenuation' : [attenuation],
        'Flocculation' : [flocculation],
        'Alcohol_Tolerance' : [abv],
        'Fermentation_Temperature' : [temp],
        'Description' : [description]}
    df = pd.DataFrame(entry)
    return df
```

```python
loadTable(soup)
```

```python
df = loadTable(soup)
df1 = df
df1
```

```python
for i in url_list:
  print(i)
  # fetch step
  html = get(i)

  # extract step
  soup = Soup(html)
  df = loadTable(soup)
  df1 = pd.concat([df1, df])
```
